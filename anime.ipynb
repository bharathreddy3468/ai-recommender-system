```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
anime = pd.read_csv('/content/drive/MyDrive/anime.csv')
anime.head()
```

```python
anime_use = anime[['MAL_ID', 'Name', 'Score', 'Genres', 'Type', 'Popularity']]
anime_use.head()
```

```python
anime_use['Type'] = anime_use['Type'].replace({'Unknown':'TV'})
```

```python
anime_use['Type'].value_counts()
```

```python
anime_use['MAL_ID'].nunique()
```

```python
anime_use['Name'].nunique()
```

```python
anime_use[anime_use.duplicated(subset=['Name'])]
```

```python
anime_use['Name'].value_counts()
```

```python
anime_use['Score'].value_counts()
```

```python
## around 30% of the column score are unknown values. Though score is important replacing that with random values or mean or median would effect the data
## drastically. Hence we will try to remove that column.

anime_use.drop(['Score'], axis=1, inplace=True)
anime_use.head()
```

```python
anime_use.drop_duplicates(subset=['Name'], inplace=True)
```

```python
anime_use.shape
```

```python
ratings = pd.read_csv('/content/drive/MyDrive/rating_complete.csv')
ratings.head()
```

```python
ratings.info()
```

```python
ratings.isnull().sum()
```

```python
ratings['user_id'].nunique()
```

```python
ratings['anime_id'].nunique()
```

```python
data = ratings.merge(anime_use, how='left', left_on='anime_id', right_on='MAL_ID')
```

```python
data.head()
```

```python
data_final = data[['user_id', 'anime_id', 'rating', 'Name', 'Type', 'Genres', 'Popularity']]
data_final.rename({'Name':'name', 'Type': 'type' ,'Genres':'genres', 'Popularity':'popularity'})
data_final.head()
```

```python
data_final.isnull().sum()
```

```python
data_final.dropna(inplace=True)
```

```python
data_final.shape
```

```python
data_final.isnull().sum()
```

```python
data_final.to_csv('/content/drive/MyDrive/anime_cleaned_2704.csv')
```
